```python
def question_1(csv_file):
    import pandas as pd

    df = pd.read_csv(csv_file)

    # 計算各縣市的大學數量
    university_counts = df.groupby("縣市名稱").size()

    # 將結果排序並輸出前五名
    return university_counts.sort_values(ascending=False).head()


def question_2(csv_file):
    import pandas as pd

    df = pd.read_csv(csv_file)

    # 計算各學校的男女學生總數
    gender_ratios = df.groupby("學校名稱").agg({"男生計": "sum", "女生計": "sum"})

    # 計算男女比例
    gender_ratios["男女比"] = gender_ratios["男生計"] / gender_ratios["女生計"]

    # 刪除不需要的欄位
    gender_ratios = gender_ratios.drop(columns=["男生計", "女生計"])

    # 按照男女比例降序排序
    gender_ratios = gender_ratios.sort_values(by="男女比", ascending=False)

    # 問題二：各學校男女比例
    print("各學校男女比例：")
    print(gender_ratios)

    # 輸出結果
    return gender_ratios




def question_3(csv_file):
    import pandas as pd

    df = pd.read_csv(csv_file)

    # 列印 "等級別" 欄位的唯一值，確認我們使用的是正確的值
    print(df["等級別"].unique())

    # 計算博士生和大學生的數量
    doctorate_count = df[df["等級別"] == "D 博士"].shape[0]
    undergraduate_count = df[df["等級別"] == "B 學士"].shape[0]

    # 列印博士生和大學生的數量，檢查是否合理
    print("博士生數量：", doctorate_count)
    print("大學生數量：", undergraduate_count)

    # 計算博士生與大學生的比例
    doctorate_undergraduate_ratio = doctorate_count / undergraduate_count

    return doctorate_undergraduate_ratio
def question_teacher(csv_file):
    import pandas as pd

    df = pd.read_csv(csv_file)



if __name__ == "__main__":
    # 設定檔案路徑
    csv_file = "./LATIA112-2/WEEK3/112_student.csv"

    # 問題一：各縣市大學數量
    print("各縣市大學數量：")
    university_counts = question_1(csv_file)
    print(university_counts.to_string())

    # 問題二：各學校男女比例
    print("各學校男女比例：")
    gender_ratios = question_2(csv_file)
    print(gender_ratios.to_string())

    # 問題三：博士生與大學生的比例
    print("博士生與大學生的比例：")
    doctorate_undergraduate_ratio = question_3(csv_file)
    print(f"{doctorate_undergraduate_ratio:.2f}")
```
