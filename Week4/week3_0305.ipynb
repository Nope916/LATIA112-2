```python
import pandas as pd
csv_file = "/content/gdrive/My Drive/LATIA112-2/WEEK3/112_student.csv"  # 請將此路徑替換為實際的檔案路徑
df = pd.read_csv(csv_file)
# 問題一的處理函式
def question_1(df):
    university_counts = df.groupby("縣市名稱").size()
    top_city_university_count = university_counts.sort_values(ascending=False).iloc[0]
    top_city = university_counts.idxmax()
    return (top_city, top_city_university_count)
# 問題二的處理函式
def question_2(df):
    gender_ratios = df.groupby("學校名稱").agg({"男生計": "sum", "女生計": "sum"})
    df['女生比例(%)'] = (df['女生計'] / (df['男生計'] + df['女生計'])) * 100
    return df
# 問題三的處理函式
def question_3(df):
    doctorate_count = df[df["等級別"] == "D 博士"].shape[0]
    undergraduate_count = df[df["等級別"] == "B 學士"].shape[0]
    doctorate_undergraduate_ratio = doctorate_count / undergraduate_count
    return doctorate_undergraduate_ratio

# 計算問題一和問題三的結果
top_city, university_counts = question_1(df)
doctorate_undergraduate_ratio = question_3(df)
# 將問題二的結果添加到DataFrame
df = question_2(df)  # 這將直接修改df，新增'女生比例(%)'欄位
# 創建一個包含問題一和問題三結果的DataFrame
summary_df = pd.DataFrame({
    "Top city by university count": [f"{top_city} with {university_counts} universities"],
    "Doctorate to undergraduate ratio": [f"{doctorate_undergraduate_ratio:.2f}"]
})

# 將原始資料的DataFrame追加到這個新的DataFrame後面
final_df = pd.concat([summary_df, df], ignore_index=True)

# 將更新後的DataFrame保存到新的CSV檔案
output_csv_file = "/content/gdrive/My Drive/LATIA112-2/WEEK3/112_students_tf.csv"
final_df.to_csv(output_csv_file, index=False)

print("Updated CSV file has been saved.")
```
